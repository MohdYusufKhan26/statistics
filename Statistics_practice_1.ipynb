# Statistics Practical Implementation

**Measure of central tendency**

1. Mean
2. Median
3. Mode


```python
ages = [23, 24, 32, 45, 12, 43, 67, 45, 32, 56, 32, 120]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

```python
import statistics  # MOde is used for missing values
statistics.mode(ages)
```

```python
import seaborn as sns

# Create box plot in x axis
sns.boxplot(x=ages)
```

# 5 Number Summary

```python
import numpy as np
q1,q3 = np.percentile(ages, [25, 75])
print(q1, q3)
```

# Check Outliers [Lower_fence, higher_fence]

```python
iqr = q3 - q1
lower_fence = q1 - 1.5*(iqr)
higher_fence = q3 + 1.5 * (iqr)
print(lower_fence, higher_fence)
```

# Measure of Dispersion

1. Variance
2. Standard deviation

```python
statistics.variance(ages)
```

```python
np.var(ages, axis = 0)
```

```python
## To calculate manually
def variance(data):
  n = len(ages)

  # mean of the data
  mean = sum(data)/n

  # variance
  deviation = [(x - mean) ** 2 for x in data]
  variance = sum(deviation) / n
  return variance

variance(ages)
```

```python
def variance(data, dof=0):
  n = len(ages)

  # mean of the data
  mean = sum(data)/n

  # variance
  deviation = [(x - mean) ** 2 for x in data]
  variance = sum(deviation) / (n - dof)
  return variance

variance(ages)
```

```python
# through statistics calculate variance
statistics.pvariance(ages)
```

Starndard deviation

```python
## standard deviation
import math
math.sqrt(statistics.pvariance(ages))
```

# Chech Whether Distribution is Normal Distribution

```python
## if you want to check whether features is guassian or normal distributed

## Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
  plt.figure(figsize = (10, 6))
  plt.subplot(1, 2, 1)
  sns.histplot(sample)
  plt.subplot(1, 2, 2)
  stat.probplot(sample, dist = 'norm', plot = pylab)
  plt.show()
```

```python
## create a normal distribute data
s = np.random.normal(0.5, 0.2, 1000)
plot_data(s)
```

```python
mu, sigma = 3., 1. #Mean and std deviation
s = np.random.lognormal(mu, sigma, 1000)
```

```python
def plot_data_log(sample):
  plt.figure(figsize = (10, 6))
  plt.subplot(1, 2, 1)
  sns.histplot(sample)
  plt.subplot(1, 2, 2)
  stat.probplot(np.log(sample), dist = 'norm', plot = pylab)
  plt.show()
```

```python
plot_data_log(s)
```
